### 등락률 계산

```python
# 컬럼별 등락률 계산

col_list = []

df = card_df[['전자상거래/통신판매', '종합소매', '자동차', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', '여행/교통', '숙박', '음식점', '공과금/개인및전문서비스', '금융/보험', '교육', '서적/문구', '의료/보건', '소비자물가지수', '소매판매지수', '국고채 3년(평균)', '국고채 5년(평균)', '국고채 10년(평균)', '회사채 3년(평균)', 'CD 91물(평균)', '콜금리(1일물,평균)', '기준금리']]
col_list.append([col for col in df])

for i in col_list:
    df[i] = df[i].pct_change()
    
df = df.iloc[1:]
df
```

### 정규성 검정

```python
from scipy import stats
from scipy.stats import shapiro

# 등락률 컬럼만 추출해서 pvalue 구하는 함수
def shap_pct(df_name):
    pct_colus = []
    pct_colus.append([col for col in df_name])
    for i in pct_colus:
        for j in i:
            shapiro_test = shapiro(df_name[j])
            print(j, "의 pvalue:", shapiro_test.pvalue)
```

```python
import numpy as np

shap_pct(df)
```

### 등락률 이상치 처리

```python
cols = []

pct_df = df[['전자상거래/통신판매', '종합소매', '자동차', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', 
                       '여행/교통', '숙박', '음식점', '서적/문구', '교육', '의료/보건', '공과금/개인및전문서비스', 
                       '금융/보험']]
cols.append([col for col in pct_df])

Q1 = pct_df[['전자상거래/통신판매', '종합소매', '자동차', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', 
                       '여행/교통', '숙박', '음식점', '서적/문구', '교육', '의료/보건', '공과금/개인및전문서비스', 
                       '금융/보험']].quantile(q=0.25)
Q3 = pct_df[['전자상거래/통신판매', '종합소매', '자동차', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', 
                       '여행/교통', '숙박', '음식점', '서적/문구', '교육', '의료/보건', '공과금/개인및전문서비스', 
                       '금융/보험']].quantile(q=0.75)
IQR = Q3-Q1

for i in cols:
    pct_df[i] = pct_df[i][(pct_df[i] <= Q3[i]+1.5*IQR[i]) & (pct_df[i] >= Q1[i]-1.5*IQR[i])]

pct_df[['스포츠/오락/여가', '식료품', '서적/문구']] = df[['스포츠/오락/여가', '식료품', '서적/문구']]
```
