```python
cols = []

card_iqr_df = card_df.copy()
card_iqr_df = card_iqr_df[['전자상거래/통신판매', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', 
                       '여행/교통', '숙박', '음식점', '서적/문구', '교육', '의료/보건', '공과금/개인및전문서비스', 
                       '금융/보험', '기타']]
cols.append([col for col in card_iqr_df])

Q1 = card_iqr_df[['전자상거래/통신판매', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', 
                       '여행/교통', '숙박', '음식점', '서적/문구', '교육', '의료/보건', '공과금/개인및전문서비스', 
                       '금융/보험', '기타']].quantile(q=0.25)
Q3 = card_iqr_df[['전자상거래/통신판매', '가구/가전', '의류/잡화', '스포츠/오락/여가', '식료품', '연료', 
                       '여행/교통', '숙박', '음식점', '서적/문구', '교육', '의료/보건', '공과금/개인및전문서비스', 
                       '금융/보험', '기타']].quantile(q=0.75)
IQR = Q3-Q1
for i in cols:
    card_iqr_df[i] = card_iqr_df[i][(card_iqr_df[i] <= Q3[i]+1.5*IQR[i]) & (card_iqr_df[i] >= Q1[i]-1.5*IQR[i])]
```

### 정규성 검정

```python
from scipy import stats
from scipy.stats import shapiro

# 등락률 컬럼만 추출해서 pvalue 구하는 함수
def shap_pct(df_name):
    pct_colus = []
    pct_colus.append([col for col in df_name])
    for i in pct_colus:
        for j in i:
            shapiro_test = shapiro(df_name[j])
            print(j, "의 pvalue:", shapiro_test.pvalue)
```

```python
import numpy as np
shap_pct(df)
```
