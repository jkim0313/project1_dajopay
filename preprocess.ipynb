```python
import pandas as pd

# csv파일 불러오기
df = pd.read_csv('./data/온라인쇼핑몰_취급상품범위별_상품군별거래액_20220804175111.csv', encoding='CP949')

df
```

```python
# 불필요 인덱스 제거

df = df.loc[1:]
```

```python
# 인덱스 재정렬

df.reset_index(drop=True, inplace=True)
df.head()
```

```python
# 데이터 타입 및 결측치 유무 확인

df.info()
```

```python
# 컬럼명 확인

df.columns
```

```python
# 컬럼명 변경

df.columns = df.columns.str.replace(".1", " 종합몰")
```

```python
# 컬럼명 변경

df.columns = df.columns.str.replace(".2", " 전문몰")
```

```python
# 컬럼명 변겅

df.columns = df.columns.str.replace("·", " ")
```

```python
# 컬럼명 변경 확인

df.columns
```

```python
# 시점 컬럼 컬럼명 변경

df.rename(columns={"시점" : "날짜"}, inplace = True)
df.head(1)
```

```python
# 날짜 컬럼 '.' 제거

df["날짜"] = df["날짜"].str.replace(".", "")
df["날짜"] = df["날짜"].str.replace(" p\)", "")
df.head(1)
```

```python
df["날짜"] = pd.to_datetime(df["날짜"], format='%Y%m')
df["날짜"] = df["날짜"].dt.strftime('%Y-%m')
df.set_index("날짜", inplace=True)
```

```python
# 날짜 컬럼을 제외한 나머지 컬럼 데이터 타입 int로 변환

df.iloc[:, 1:] = df.iloc[:, 1:].astype(int)
df.info()
```

```python
# 특정 단어를 포함하는 컬럼 제거

df = df[df.columns.drop(df.filter(regex="종합몰"))]
df = df[df.columns.drop(df.filter(regex="전문몰"))]
df.head(1)
```

```python
df.columns
```

```python
df_food = df["음 식료품"]

df_food.head()
```

```python
cpi_df = pd.read_csv('./data/소비자물가지수_2020100__20220803183830.csv', encoding='CP949')
cpi_df
```

```python
cpi_df = cpi_df.iloc[:-1]
cpi_df
```

```python
cpi_df.info()
```

```python
cpi_df["시점"] = cpi_df["시점"].astype(str)
```

```python
# .10 이 .1로 변환된 걸 다시 재변환

cpi_df["시점"][cpi_df["시점"].str.endswith(".1")] = cpi_df["시점"][cpi_df["시점"].str.endswith(".1")]+"0"
```

```python
cpi_df
```

```python
# 날짜로 형변환 후 인덱스로 설정

cpi_df["시점"] = pd.to_datetime(cpi_df["시점"], format="%Y.%m")
cpi_df["시점"] = cpi_df["시점"].dt.strftime('%Y-%m')
cpi_df.set_index("시점", inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["font.family"] = "Malgun Gothic"


plt.subplots(figsize=(12, 10), constrained_layout=True)
plt.subplot(2, 1, 1)
plt.title("음 식료품")
plt.plot(df_food, c='b', label='음 식료품')
plt.legend()
plt.xticks(rotation=90) 
plt.grid(True)  

plt.subplot(2, 1, 2)
plt.title("소비자 물가 지수")
plt.plot(cpi_df, c='r', label='CPI')
plt.legend()
plt.xticks(rotation=90)         # 텍스트 90도 회전
plt.grid(True)                  # 그리드 표시

plt.show()
```
