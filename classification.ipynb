```python
# 서적/문구 데이터프레임 복사
month_df = book_df.copy()
month_df = month_df[["날짜", "전자상거래/통신판매"]]
month_df.head()
```

```python
# 동월 전체 평균
month_df["전자상거래/통신판매"][month_df["날짜"].str.endswith("12")].mean()
```

```python
month_df.head()
```

```python
pre_mean = month_df["전자상거래/통신판매"].mean()
```

```python
# 각 동월 전체 평균 리스트 저장

month_list = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10", "11", "12"]
month_mean = []
for i in month_list:
    print(month_df["전자상거래/통신판매"][month_df["날짜"].str.endswith(i)].mean())
    month_mean.append(month_df["전자상거래/통신판매"][month_df["날짜"].str.endswith(i)].mean())
    
month_mean        
```

```python
# "동월_평균_등락률" 컬럼 생성 및 값 저장

month_df["동월_평균_등락률"] = ""
for i in month_list:
        month_df["동월_평균_등락률"][month_df["날짜"].str.endswith(i)]  = month_df["전자상거래/통신판매"][month_df["날짜"].str.endswith(i)].mean()
month_df.head(20)
```

```python
len(month_df)
```

```python
# "전체_평균_대비" 컬럼 생성 및 값 저장

month_df["전체_평균_대비"] = ""

# 동월 평균 등락률이 전체 평균 보다 높거나 같으면 1, 낮으면 0
for i in month_df.index:
    if month_df.loc[i, "동월_평균_등락률"] >= pre_mean:
        month_df.loc[i, "전체_평균_대비"] = 1
    else:
        month_df.loc[i, "전체_평균_대비"] = 0

month_df.tail(20)
```
