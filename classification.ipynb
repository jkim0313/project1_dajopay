```python
import numpy as np

book_df["날짜"] = book_df["날짜"].astype(str)
book_df["소비"] = ""

for i in range(0, 13):
    for j in range(12, 85):
        # 첫 1년
        if book_df.loc[i, "날짜"].endswith("03"):
            book_df.loc[i, "소비"] = 1
        else:
            if book_df.loc[i,"서적/문구"] > 0: 
                if book_df.loc[i,"소매판매지수"] >= book_df["소매판매지수"].quantile(q=0.5):
                    book_df.loc[i,"소비"] = 1
                elif book_df.loc[i,"소매판매지수"] >= book_df["소매판매지수"].quantile(q=0.25):
                    if book_df.loc[i,"소비자물가지수"] <= book_df["소비자물가지수"].quantile(q=0.75):
                        book_df.loc[i,"소비"] = 1
                    else:
                        book_df.loc[i,"소비"] = 0
                elif book_df.loc[i,"소매판매지수"] < book_df["소매판매지수"].quantile(q=0.25):
                    book_df.loc[i,"소비"] = 0

            elif book_df.loc[i, "서적/문구"] <= 0:
                if book_df.loc[i,"소매판매지수"] <= book_df["소매판매지수"].quantile(q=0.75):
                    book_df.loc[i,"소비"] = 0
                elif book_df.loc[i,"소매판매지수"] > book_df["소매판매지수"].quantile(q=0.75):
                    if book_df.loc[i,"소비자물가지수"] > book_df["소비자물가지수"].quantile(q=0.5):
                        book_df.loc[i,"소비"] = 0
                    else:
                        book_df.loc[i,"소비"] = 1

        # 첫 1년 이후
        if book_df["서적/문구"].loc[j] > 0:
            if book_df["서적/문구"].loc[j] > book_df["서적/문구"].loc[j-12]:
                book_df.loc[j,"소비"] = 1
            elif book_df["서적/문구"].loc[j] <= book_df["서적/문구"].loc[j-12]:
                if book_df.loc[j,"소매판매지수"] > book_df["소매판매지수"].quantile(q=0.25): 
                    if book_df.loc[j,"소비자물가지수"] < book_df["소비자물가지수"].quantile(q=0.75):
                        book_df.loc[j,"소비"] = 1
                    else:
                        book_df.loc[j,"소비"] = 0
                else:
                    book_df.loc[j,"소비"] = 0
            
        elif book_df["서적/문구"].loc[j] <= 0:
            if book_df["서적/문구"].loc[j] <= book_df["서적/문구"].loc[j-12]:
                if book_df.loc[j,"소매판매지수"] <= book_df["소매판매지수"].quantile(q=0.75): 
                    book_df.loc[j, "소비"] = 0
                elif book_df.loc[j,"소매판매지수"] > book_df["소매판매지수"].quantile(q=0.75):
                    if book_df.loc[j,"소비자물가지수"] > book_df["소비자물가지수"].quantile(q=0.5):
                        book_df.loc[j,"소비"] = 0
                    else:
                        book_df.loc[j,"소비"] = 1
            elif book_df["서적/문구"].loc[j] > book_df["서적/문구"].loc[j-12]:
                if book_df.loc[j,"소매판매지수"] > book_df["소매판매지수"].quantile(q=0.75):
                    book_df.loc[j,"소비"] = 1
                else:
                    book_df.loc[j,"소비"] = 0
            
book_df
```
