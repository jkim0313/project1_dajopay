```python
import numpy as np

book_df["날짜"] = book_df["날짜"].astype(str)
book_df["소비"] = ""

for i in range(0, 13):
    for j in range(12, 85):
        if book_df.loc[i,"서적/문구"] > 0: 
            if book_df.loc[i,"소비자물가지수"] <= book_df["소비자물가지수"].quantile(q=0.75): 
                if book_df.loc[i,"소매판매지수"] >= book_df["소매판매지수"].quantile(q=0.25):
                    book_df.loc[i,"소비"] = 1
            elif book_df.loc[i,"소비자물가지수"] > book_df["소비자물가지수"].quantile(q=0.75):
                if book_df.loc[i,"소매판매지수"] >= book_df["소매판매지수"].quantile(q=0.5):
                    book_df.loc[i,"소비"] = 1
            elif book_df.loc[i,"소비자물가지수"] > book_df["소비자물가지수"].quantile(q=0.75):
                if book_df.loc[i,"소매판매지수"] < book_df["소매판매지수"].quantile(q=0.5):
                    book_df.loc[i,"소비"] = 0

        elif book_df.loc[i, "서적/문구"] <= 0:
            if book_df.loc[i,"소비자물가지수"] > book_df["소비자물가지수"].quantile(q=0.25): 
                if book_df.loc[i,"소매판매지수"] < book_df["소매판매지수"].quantile(q=0.75):
                    book_df.loc[i,"소비"] = 0
                else:
                    book_df.loc[i,"소비"] = 1
            elif book_df.loc[i,"소비자물가지수"] <= book_df["소비자물가지수"].quantile(q=0.25):
                if book_df.loc[i,"소매판매지수"] > book_df["소매판매지수"].quantile(q=0.75):
                    book_df.loc[i,"소비"] = 1
                else:
                    book_df.loc[i,"소비"] = 0


        if book_df["서적/문구"].loc[j] > 0:
            if book_df["서적/문구"].loc[j-12] > 0:
                if book_df["서적/문구"].loc[j] > book_df["서적/문구"].loc[j-12]:
                    book_df.loc[j,"소비"] = 1
                elif book_df["서적/문구"].loc[j] < book_df["서적/문구"].loc[j-12]:
                    if (book_df["서적/문구"].loc[j]) > (book_df["서적/문구"].loc[j-12]/(1/2)):                        
                        if book_df.loc[j,"소비자물가지수"] <= 0.001091: 
                            if book_df.loc[j,"소매판매지수"] >= -0.003923:
                                book_df.loc[j,"소비"] = 1
                        elif book_df.loc[j,"소비자물가지수"] > 0.001091:
                            if book_df.loc[j,"소매판매지수"] >= 0.009050:
                                book_df.loc[j,"소비"] = 1
                        elif book_df.loc[j,"소비자물가지수"] > 0.001091:
                            if book_df.loc[j,"소매판매지수"] < 0.009050:
                                book_df.loc[j,"소비"] = 0 
                    
        elif book_df["서적/문구"].loc[j] <= 0:
            if book_df["서적/문구"].loc[j] > book_df["서적/문구"].loc[j-12]:
                if book_df.loc[j, "서적/문구"] > -0.024028:
                    if book_df.loc[j,"소비자물가지수"] <= np.mean(book_df["소비자물가지수"]): 
                        if book_df.loc[j,"소매판매지수"] >= np.mean(book_df["소매판매지수"]):
                            book_df.loc[j,"소비"] = 1
                    elif book_df.loc[j,"소비자물가지수"] > np.mean(book_df["소비자물가지수"]):
                        if book_df.loc[j,"소매판매지수"] >= np.mean(book_df["소매판매지수"]):
                            book_df.loc[j,"소비"] = 1
                    elif book_df.loc[j,"소비자물가지수"] > np.mean(book_df["소비자물가지수"]):
                        if book_df.loc[j,"소매판매지수"] < np.mean(book_df["소매판매지수"]):    
                            book_df.loc[j,"소비"] = 0
            elif book_df["서적/문구"].loc[j] < book_df["서적/문구"].loc[j-12]:
                 book_df.loc[j,"소비"] = 0
book_df
```

```python
import numpy as np

book_df["날짜"] = book_df["날짜"].astype(str)
book_df["소비"] = ""

for i in book_df.index:
    # 양수는 일단 1을 준다
    (book_df["서적/문구"].loc[12] > 0)
    (book_df["서적/문구"].loc[12-12] > 0)
    # 전년 동월보다 크고
    (book_df["서적/문구"].loc[12] > book_df["서적/문구"].loc[12-12])
    # 전년 동월보다 작아도 양수이고 전년 동월값의 50% 이상이면? 
    (book_df["서적/문구"].loc[12] < book_df["서적/문구"].loc[12-12])
    (book_df["서적/문구"].loc[12] > book_df["서적/문구"].loc[12-12])/(1/2) 


    # 음수지만 작년 동월보다 높고
    (book_df["서적/문구"].loc[12] <= 0) 
    (book_df["서적/문구"].loc[12] > book_df["서적/문구"].loc[12-12])

    # 작년 동월보다 낮으면 0
    (book_df["서적/문구"].loc[12] < book_df["서적/문구"].loc[12-12])
    
    
    if book_df.loc[i,"점수"]==1:
        if book_df.loc[i,"소비자물가지수"] <= 0.001091: 
            if book_df.loc[i,"소매판매지수"] >= 0.009050:
                book_df.loc[i,"소비"] = 1
        elif book_df.loc[i,"소비자물가지수"] > 0.001091:
            if book_df.loc[i,"소매판매지수"] >= 0.009050:
                book_df.loc[i,"소비"] = 1
        elif book_df.loc[i,"소비자물가지수"] > 0.001091:
            if book_df.loc[i,"소매판매지수"] < 0.009050:
                book_df.loc[i,"소비"] = 0 
                
    elif book_df.loc[i,"점수"]==0:
        if book_df.loc[i, "서적/문구"] > -0.024028:
            if book_df.loc[i,"소비자물가지수"] <= np.mean(book_df["소비자물가지수"]): 
                if book_df.loc[i,"소매판매지수"] >= np.mean(book_df["소매판매지수"]):
                    book_df.loc[i,"소비"] = 1
            elif book_df.loc[i,"소비자물가지수"] > np.mean(book_df["소비자물가지수"]):
                if book_df.loc[i,"소매판매지수"] >= np.mean(book_df["소매판매지수"]):
                    book_df.loc[i,"소비"] = 1
            elif book_df.loc[i,"소비자물가지수"] > np.mean(book_df["소비자물가지수"]):
                if book_df.loc[i,"소매판매지수"] < np.mean(book_df["소매판매지수"]):    
                    book_df.loc[i,"소비"] = 0 
    # else:
    #     book_df.loc[i,"소비"] = 0
book_df
```
