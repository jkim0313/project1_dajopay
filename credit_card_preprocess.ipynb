### 개인 신용카드 승인금액

```python
import pandas as pd 

# 모든 컬럼 보기
pd.set_option('display.max_columns', None)

credit_df = pd.read_csv('./data/지역_유형별 신용카드_(단위_억).csv')
credit_df.head()
```

```python
credit_df.columns
```

```python
# 날짜 컬럼 데이터타입 확인
credit_df["지역, 카테고리"].info()
```

```python
credit_df.rename(columns={"지역, 카테고리" : "날짜"}, inplace=True)
```

```python
# 날짜 컬럼 "/" 제거
credit_df["날짜"] = credit_df["날짜"].str.replace("/", "")
credit_df.head()
```

```python
# 상품군 나누기
total_acco_res_df = credit_sum_df[['날짜', '전국_숙박/음식', '전국_숙박', '전국_음식점']]
total_shopping_df = credit_sum_df[['날짜', '전국_백화점', '전국_대형마트/유통전문점', '전국_슈퍼마켓', '전국_편의점']]
total_vehicle_df = credit_sum_df[['날짜', '전국_자동차', '전국_국산자동차신품', '전국_기타운송수단', '전국_자동차 부품 및 정비', '전국_연료']]
```

#### 계절별 나누기

```python
# 계절 나누는 함수
def seasons(s):
    if s.endswith("3") or s.endswith("4") or s.endswith("5"):
        return "봄"
    elif s.endswith("6") or s.endswith("7") or s.endswith("8"):
        return "여름"
    elif s.endswith("9") or s.endswith("10") or s.endswith("11"):
        return "가을"
    else:
        return "겨울"
```

```python
# str 타입이 아니면 함수를 적용 못하기 때문에 데이터 타입 변경
total_acco_pct_df["날짜"] = total_acco_pct_df["날짜"].astype(str)
```

```python
# 계졀 나누는 함수 적용
total_acco_pct_df["계절"] = total_acco_pct_df["날짜"].apply(seasons)
total_shopping_pct_df["계절"] = total_shopping_pct_df["날짜"].apply(seasons)
total_vehicle_pct_df["계절"] = total_vehicle_pct_df["날짜"].apply(seasons)
total_acco_pct_df.head()
```

```python
# 계절 컬럼에 년도 추가
total_acco_pct_df["계절"] = total_acco_pct_df["날짜"].str[:4] + " " + total_acco_pct_df["계절"]
total_shopping_pct_df["계절"] = total_shopping_pct_df["날짜"].str[:4] + " " + total_shopping_pct_df["계절"]
total_vehicle_pct_df["계절"] = total_vehicle_pct_df["날짜"].str[:4] + " " + total_vehicle_pct_df["계절"]
total_acco_pct_df.head()
```

```python
from scipy.stats import shapiro

# 등락률 컬럼만 추출해서 pvalue 구하는 함수
def shap_pct(df_name):
    pct_colus = []
    pct_colus.append([col for col in df_name if "등락률" in col])
    for i in pct_colus:
        for j in i:
            shapiro_test = shapiro(df_name[j])
            print(j, "의 pvalue:", shapiro_test.pvalue)
```

```python
shap_pct(total_vehicle_pct_df)
```

```python
# 등락률 컬럼만 추출해서 표준편차 구하는 함수

def std_pct(df_name):
    pct_colus = []
    pct_colus.append([col for col in df_name if "등락률" in col])
    for i in pct_colus:
        print("컬럼별 표준편차:\n", df_name[i].std(ddof=0))
```

```python
std_pct(total_acco_pct_df)
```
