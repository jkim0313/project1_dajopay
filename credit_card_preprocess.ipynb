### 개인 신용카드 승인금액

```python
import pandas as pd 

# 모든 컬럼 보기
pd.set_option('display.max_columns', None)

credit_df = pd.read_csv('./data/지역별 소비유형별 개인 신용카드 승인금액_1601_2204.csv')
credit_df.head()
```

```python
credit_df.columns
```

```python
# 날짜 컬럼 데이터타입 확인
credit_df["날짜"].info()
```

```python
# 날짜 컬럼 "/" 제거
credit_df["날짜"] = credit_df["날짜"].str.replace("/", "")
credit_df.head()
```

```python
# 날짜 컬럼 int 타입으로 변환
credit_df["날짜"] = credit_df["날짜"].astype(int)
```

```python
# 날짜 컬럼 201710 이후 날짜만 남긴다.
credit_df = credit_df[credit_df["날짜"] > 201710]
credit_df.reset_index(drop=True, inplace=True)
```

```python
credit_df.head()
```

```python
credit_sum_df = credit_df.iloc[:, credit_df.columns.str.contains("전국")]
credit_sum_df.columns
```

```python
credit_sum_df = pd.DataFrame(credit_df, columns=[
        '날짜', '전국_합계', '전국_종합소매', '전국_백화점', '전국_대형마트/유통전문점', '전국_슈퍼마켓', '전국_편의점',
       '전국_면세점', '전국_전자상거래/통신판매', '전국_식료품', '전국_일반식료품', '전국_건강보조식품',
       '전국_의류/잡화', '전국_의복/직물', '전국_복식잡화', '전국_시계/귀금속/안경', '전국_화장품', '전국_연료',
       '전국_가구/가전', '전국_가구', '전국_가전제품/정보통신기기', '전국_의료/보건', '전국_종합병원',
       '전국_일반병의원/기타의료기관', '전국_자동차', '전국_국산자동차신품', '전국_기타운송수단',
       '전국_자동차 부품 및 정비', '전국_여행/교통', '전국_여행사/자동차임대', '전국_항공사', '전국_대중교통',
       '전국_오락/문화', '전국_스포츠/오락/여가', '전국_서적/문구', '전국_교육', '전국_숙박/음식', '전국_숙박',
       '전국_음식점', '전국_공과금/개인 및 전문 서비스', '전국_금융/보험', '전국_기타'])
credit_sum_df.head()
```

```python
# 전국_항공사 데이터 타입 object 확인

credit_sum_df.info()
```

```python
# 괄호 제거 및 int64로 변경
credit_sum_df["전국_항공사"] = credit_sum_df["전국_항공사"].str.replace("\(", "-")
credit_sum_df["전국_항공사"] = credit_sum_df["전국_항공사"].str.replace("\)", "")
credit_sum_df["전국_항공사"] = credit_sum_df["전국_항공사"].astype('int64')
```

```python
# 억단위만 남기기
credit_sum_df.iloc[:, 1:] = (credit_sum_df.iloc[:, 1:]/1000000)
credit_sum_df.iloc[:, 1:] = credit_sum_df.iloc[:, 1:].astype(int)
credit_sum_df.head()
```

```python
# 상품군 나누기
total_acco_res_df = credit_sum_df[['날짜', '전국_숙박/음식', '전국_숙박', '전국_음식점']]
total_shopping_df = credit_sum_df[['날짜', '전국_백화점', '전국_대형마트/유통전문점', '전국_슈퍼마켓', '전국_편의점']]
total_vehicle_df = credit_sum_df[['날짜', '전국_자동차', '전국_국산자동차신품', '전국_기타운송수단', '전국_자동차 부품 및 정비', '전국_연료']]
```

```python
# 등락률 컬럼 추가
# 숙박/음식
total_acco_res_df['전국_숙박/음식_등락률'] = total_acco_res_df['전국_숙박/음식'].pct_change()
total_acco_res_df['전국_숙박_등락률'] = total_acco_res_df['전국_숙박'].pct_change()
total_acco_res_df['전국_음식점_등락률'] = total_acco_res_df['전국_음식점'].pct_change()

# 쇼핑
total_shopping_df['전국_백화점_등락률'] = total_shopping_df['전국_백화점'].pct_change()
total_shopping_df['전국_대형마트/유통전문점_등락률'] = total_shopping_df['전국_대형마트/유통전문점'].pct_change()
total_shopping_df['전국_슈퍼마켓_등락률'] = total_shopping_df['전국_슈퍼마켓'].pct_change()
total_shopping_df['전국_편의점_등락률'] = total_shopping_df['전국_편의점'].pct_change()

# 자동차
total_vehicle_df['전국_자동차_등락률'] = total_vehicle_df['전국_자동차'].pct_change()
total_vehicle_df['전국_국산자동차신품_등락률'] = total_vehicle_df['전국_국산자동차신품'].pct_change()
total_vehicle_df['전국_기타운송수단_등락률'] = total_vehicle_df['전국_기타운송수단'].pct_change()
total_vehicle_df['전국_자동차 부품 및 정비_등락률'] = total_vehicle_df['전국_자동차 부품 및 정비'].pct_change()
total_vehicle_df['전국_연료_등락률'] = total_vehicle_df['전국_연료'].pct_change()
```

```python
total_acco_res_df.dropna(inplace=True)
total_acco_res_df["날짜"] = pd.to_datetime(total_acco_res_df["날짜"], format='%Y%m')
total_acco_res_df["날짜"] = total_acco_res_df["날짜"].dt.strftime('%Y%m')
# total_acco_res_df.set_index("날짜", inplace=True)
total_acco_res_df = total_acco_res_df.iloc[:-2]
total_acco_res_df.head(2)
```

### 소비자 물가 지수(CPI)

```python
cpi_df = pd.read_csv('./data/소비자물가지수_2020100__20220808154831.csv', encoding='CP949')
cpi_df
```

```python
cpi_df["CPI_등락률"] = cpi_df["전국"].pct_change()
```

```python
cpi_df = cpi_df.iloc[1:-5]
cpi_df.tail(2)
```

```python
cpi_df["시점"] = cpi_df["시점"].astype(str)

# .10 이 .1로 변환된 걸 다시 재변환

cpi_df["시점"][cpi_df["시점"].str.endswith(".1")] = cpi_df["시점"][cpi_df["시점"].str.endswith(".1")]+"0"
```

```python
# 날짜로 형변환 후 인덱스로 설정

cpi_df["시점"] = pd.to_datetime(cpi_df["시점"], format="%Y.%m")
cpi_df["시점"] = cpi_df["시점"].dt.strftime('%Y-%m')
cpi_df.rename(columns={"시점" : "날짜"}, inplace=True)
# cpi_df.set_index("날짜", inplace=True)
```

#### 계절별 나누기

```python
# 계절 나누는 함수
def seasons(s):
    if s.endswith("3") or s.endswith("4") or s.endswith("5"):
        return "봄"
    elif s.endswith("6") or s.endswith("7") or s.endswith("8"):
        return "여름"
    elif s.endswith("9") or s.endswith("10") or s.endswith("11"):
        return "가을"
    else:
        return "겨울"
```

```python
# str 타입이 아니면 함수를 적용 못하기 때문에 데이터 타입 변경
total_acco_cpi_df["날짜"] = total_acco_cpi_df["날짜"].astype(str)
```
